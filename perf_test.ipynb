```python
from bin_model import BinomialModel, get_up_factor, get_down_factor
from options import EuropeanCall
import timeit
```

```python
dt = 1 / (12 * 30 * 2)
sigma = 0.3
u = get_up_factor(sigma, dt)
d = get_down_factor(sigma, dt)
S0 = 50
r = 0.02
K = 48
T = 2

N = int(T / dt) # number of periods
bin_model = BinomialModel(S0, N, dt, u, d, r)
```

```python
print(f"Number of periods: {N}")
```

```python
eur_call = EuropeanCall(K)

execution_time = timeit.timeit(
    lambda: eur_call.price(bin_model),
    number=1
)

print(f"Total time: {execution_time:.6f} seconds")
print(f"Average time: {execution_time / 1:.10f} seconds")
```
